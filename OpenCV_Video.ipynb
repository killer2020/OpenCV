```python
cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Display the resulting frame
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
#Captures videp flips it and saves
#issue faced: if we use some other resolution
# then macbook normal resolution(1280,720) doest work 

import numpy as np
import cv2

cap = cv2.VideoCapture(0)

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('flippedVideo.avi',fourcc, 20.0, (1280,720))
#cv2.startWindowThread()

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        frame = cv2.flip(frame,0)

        # write the flipped frame
        out.write(frame)

        cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            print('Video exit')
            break
    else:
        break

# Release everything if job is finished
print('start release cp#1')
cap.release()
print('start release cp#2')
out.release()
print('start release cp#3')
cv2.destroyAllWindows()
cv2.waitKey(1)
print('start release cp#4')
```
